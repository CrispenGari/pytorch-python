```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random 
import json
```

```python
class Electronics:
    RAW_IMAGES = 'data'
    ROOT = "electronics"
    
    CAMERA = "camera"
    CELLPHONE = 'cellphone'
    CHARGER = 'charger'
    DESKTOP = 'desktop'
    DRONE = 'drone'
    HEADPHONE = 'headphone'
    LAPTOP = 'laptop'
    MOUSE = 'mouse'
    TELEVISION = 'television'
    REMOTE = 'remote'
    
    GRAY = 'gray'
    BGR = 'bgr'
    RGB = 'rgb'
```

```python
class_names = dict([(key, value) for (value, key ) in enumerate(os.listdir(Electronics.RAW_IMAGES))])
class_names
```

```python
def imageGenerator(catetory:str, size:int = max_len):
    images_list = os.listdir(os.path.join(Electronics.RAW_IMAGES, category))
    random.shuffle(images_list)
    for img_path in images_list[:size]:
        WIDTH, HEIGHT = 100, 100
        """
        Saving paths
        """
        image_path_gray = os.path.join(os.path.join(os.path.join(
            Electronics.ROOT, category), Electronics.GRAY))
        image_path_rgb = os.path.join(os.path.join(os.path.join(
            Electronics.ROOT, category), Electronics.RGB))
        image_path_bgr = os.path.join(os.path.join(os.path.join(
            Electronics.ROOT, category), Electronics.BGR))

        if os.path.exists(image_path_gray) == False:
            os.makedirs(image_path_gray)
        if os.path.exists(image_path_bgr) == False:
            os.makedirs(image_path_bgr)
        if os.path.exists(image_path_rgb) == False:
            os.makedirs(image_path_rgb)
        """
        Manipulating pixels
        """
        image_path = os.path.join(os.path.join(Electronics.RAW_IMAGES, category), img_path)
        image_bgr = cv2.resize(cv2.imread(image_path), (WIDTH, HEIGHT))
        image_gray = cv2.resize(cv2.imread(image_path, cv2.IMREAD_GRAYSCALE), (WIDTH, HEIGHT))
        image_rgb = cv2.resize(cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB), (WIDTH, HEIGHT))

        """
        Save the manipulated images.
        """
        print(f"Saving {img_path}...")
        cv2.imwrite(f'{image_path_rgb}\\{img_path}' , image_rgb)
        cv2.imwrite(f'{image_path_bgr}\\{img_path}', image_bgr)
        cv2.imwrite(f'{image_path_gray}\\{img_path}', image_gray)
    print(f"\nSave History\nTotal:\t\t {max_len}\nCategory:\t {catetory.title()}\n\n")
```

> Generating ``images``.

```python
categories = np.array([
    Electronics.CAMERA,
    Electronics.CELLPHONE,
    Electronics.CHARGER,
    Electronics.DESKTOP,
    Electronics.DRONE,
    Electronics.HEADPHONE,
    Electronics.LAPTOP,
    Electronics.MOUSE,
    Electronics.TELEVISION,
    Electronics.REMOTE,
])
categories
```

```python
for category in categories:
    imageGenerator(category)
with open(f'{Electronics.ROOT}\\class_names.json', 'w') as f:
    json.dump(class_names, f)
print("Saved")
```

### Image Cleaning
> We want to prepare our images to have the same shape. For this task im going to create 2 folders of gray images and colored images of the same shape.

```python
import os, cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
class Images:
    GRAY_IMAGES_BEES = "data/gray/bees"
    GRAY_IMAGES_ANTS = "data/gray/ants"
    COLORED_IMAGES_ANTS_RGB = 'data/colored/rgb/ants'
    COLORED_IMAGES_BEES_RGB = "data/colored/rgb/bees"
    COLORED_IMAGES_ANTS_BGR = 'data/colored/bgr/ants'
    COLORED_IMAGES_BEES_BGR = "data/colored/bgr/bees"
    BEES_PATH = 'data/bees_raw'
    ANTS_PATH = 'data/ants_raw'
```

```python
def convert(path, gray, colored_rgb, colored_bgr):
    paths = os.listdir(path)
    for p in paths:
        image = cv2.imread(os.path.join(path, p))
        image = cv2.resize(image, (200, 200))
        image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        
        if not os.path.exists(gray):
            os.makedirs(gray)
        if not os.path.exists(colored_rgb):
            os.makedirs(colored_rgb)
        if not os.path.exists(colored_bgr):
            os.makedirs(colored_bgr)
        print("Writting image: ", p)
        cv2.imwrite(os.path.join(gray,p), image_gray)
        cv2.imwrite(os.path.join(colored_rgb,p), image_rgb)
        cv2.imwrite(os.path.join(colored_bgr,p), image)
convert(Images.BEES_PATH, Images.GRAY_IMAGES_BEES, Images.COLORED_IMAGES_BEES_RGB, Images.COLORED_IMAGES_BEES_BGR)
convert(Images.ANTS_PATH, Images.GRAY_IMAGES_ANTS, Images.COLORED_IMAGES_ANTS_RGB, Images.COLORED_IMAGES_ANTS_BGR)
```

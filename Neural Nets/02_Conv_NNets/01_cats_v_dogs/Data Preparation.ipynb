### Cats vrs Dogs
> Im going to prepare the data to train my `Conv NNet` using the data that i've already downladed from my previous projects `cats-vrs-dogs` using `sklearn-image-classification`. There are 2 folders `cats_gray` and `dogs_gray` which contains 100 images of `100x100` grayscale images. I've already moved these files in the root directory.

### Imports

```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
dogs_path = 'dogs_gray'
cats_path = 'cats_gray'
```

```python
le = LabelEncoder()
target_names = np.array(["cat", "dog"])
target = np.array(le.fit_transform(target_names))
```

> `0` for cat and `1` for dog

```python
data = []
```

### Cats

```python
for file in os.listdir(cats_path):
    image = np.array(cv2.imread(os.path.join(cats_path, file)))
    image = image/255
    image = image[:, :, 0] # drop the last channel
    data.append([image, np.eye(2)[0]])
```

### Dogs

```python
for file in os.listdir(dogs_path):
    image = np.array(cv2.imread(os.path.join(dogs_path, file)))
    image = image/255
    image = image[:, :, 0] # drop the last channel
    data.append([image, np.eye(2)[1]])
```

### Shuffling the data

```python
np.random.shuffle(data)
```

### Saving the data

```python
np.save('cats_v_dogs.npy', data)
print("SAVED!!")
```

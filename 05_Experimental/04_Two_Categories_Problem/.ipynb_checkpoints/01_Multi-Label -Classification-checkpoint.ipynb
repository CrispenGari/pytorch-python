### Mutiple Labels Classification

> Predict a color and a marker at the same time.
[Credits](https://discuss.pytorch.org/t/a-model-with-multiple-outputs/10440/7)

### Imports

```python
import torch
from torch.utils.data import DataLoader, Dataset
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from torch import nn
from torch.nn import functional as F
import torch.optim as optim
import pandas as pd
from collections import Counter
```

### Creating Datasets

```python
class Colors(Dataset):
    def __init__(self,path, transform=None, shuffle=False):
        self.path = path
        self.transform = transform
        self.dataFrame = pd.read_csv(path)
        self.xy = pd.read_csv(path).values
        self.shuffle = shuffle
        
        if self.shuffle:
            np.random.shuffle(self.xy)
            
        self.x = self.xy[:, :2]
        self.color_cate = self.xy [:,-2:-1].reshape(-1)
        self.marker_cate = self.xy[:, -1]
        self.colors_dict = dict([(value, key) for (key, value) in enumerate(['red', 'blue', 'green', 'teal', 'orange', 'purple'])])
        self.marker_dict = dict([(value, key) for (key, value) in enumerate(['^', '+', '*'])])
        self.y_colors = np.array([self.colors_dict[i] for i in self.color_cate])
        self.y_markers = np.array([self.marker_dict[i] for i in self.marker_cate])
        print( self.y_colors, self.y_markers)

        self.len = self.y_colors.shape[0]
        
    def __len__(self):
        return self.len
    
    def __getitem__(self, index):
        sample = self.x[index], self.y_colors[index], self.y_markers[index]
        if self.transform:
            sample = self.transform(sample)
        return sample
    
    def plot(self):
        sns.scatterplot(data=self.dataFrame, x='x', y='y',hue='color')
        plt.show()
```

### Creating a Transformers

```python
class ToTensor:
    def __call__(self, samples):
        x, y_colors, y_markers = samples
        return torch.from_numpy(x.astype('float32')) ,torch.from_numpy(np.array(y_colors)).long(),torch.from_numpy(np.array(y_markers)).long()
```

```python
train = Colors(
    'data/train.csv',
    transform =ToTensor(),
    shuffle=True
)
test = Colors(
    'data/test.csv',
    ToTensor(),
    shuffle=True
)
```

```python
train[:2]
```

### LOading the data using the DataLoader

```python
train_set = DataLoader(dataset=train,
                      batch_size =10,
                      shuffle=True)
test_set = DataLoader(dataset=test,
                      batch_size =10,
                      shuffle=False)
```

### Creating a Model

```python
class Net(nn.Module):
    def __init__(self):
        super().__init__()
        # for the color
        self.fc1 = nn.Linear(2, 16)
        self.fc2 = nn.Linear(16, 16)
        self.fc3 = nn.Linear(16, 6)
        # for the marker
        self.fc4 = nn.Linear(2, 16)
        self.fc5 = nn.Linear(16, 16)
        self.fc6 = nn.Linear(16, 1)
    
    def forward(self, x):
        y = x
        # color
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x =  self.fc3(x)
        
        # marker
        y = F.relu(self.fc4(y))
        y = F.relu(self.fc5(y))
        y =  torch.sigmoid(self.fc6(y))
        return x, y
net = Net()
net
```

```python
criterion1 = torch.nn.CrossEntropyLoss()
criterion2 = torch.nn.BCELoss()
optimizer = torch.optim.Adam(net.parameters(), lr=0.01)
```

```python
EPOCHS = 5
for epoch in range(EPOCHS):
    for X, y1, y2  in train_set:
        optimizer.zero_grad()
        # forward pass
        out1, out2 = net(X)
        # print(out2, y1, y2, out1)
        loss1 = criterion1(out1, y1)
        loss2 = criterion2(out2, y2.type(torch.FloatTensor).unsqueeze_(1))
        loss = loss1 + loss2
        # backward pass
        loss.backward()
        # update the weights
        optimizer.step()
        
    if (epoch + 1) % 1 == 0:
        print(f'Epoch {epoch+1}/{EPOCHS}')
        print(f"loss: {loss.item():.3f}, accuracy: 0"))
```

### Evaluation

```python
total_colors = 0
correct_colors = 0
total_markers = 0
correct_markers = 0
with torch.no_grad():
    for data in test_set:
        X, y1, y2 = data
        output1, output2 = net(X)

        output1 = torch.argmax(output1, dim=1)
        output2 = torch.round_(output2.squeeze_(-1))
        for i in range(len(output1)):
            if output1[i] == y1[i]:
                correct_colors +=1
            total_colors +=1
            
        for i in range(len(output2)):
            if output2[i] == y2[i]:
                correct_markers +=1
            total_markers +=1
            
    print(f"Color Accuracy: {correct_colors/total_colors:.3f}, Marker Accuracy: {correct_markers/total_markers:.3f}")
```
